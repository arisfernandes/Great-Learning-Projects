```python
# Create a new DataFrame by selecting only the desired columns for analysis 
data = Zomoto_dataset[['Unnamed: 0', 'name', 'online_order', 'book_table','rate', 'votes', 'listed_in(city)', 'location', 
      'rest_type', 'cuisines', 'approx_cost(for two people)', 'listed_in(type)']]

# Rename the columns with appropriate names 
data.rename({'Unnamed: 0': 'Number', 'name': 'Name', 'online_order': 'Online Order', 'book_table': 'Table Reservation',
             'rate': 'Ratings', 'votes': 'Votes', 'location': 'Location', 'rest_type': 'Restro Type', 
             'cuisines': 'Cuisines', 'approx_cost(for two people)': 'Cost', 'listed_in(type)': 'Type', 
             'listed_in(city)': 'City'}, axis = 1, inplace = True)

# Retrieve first five observations from the dataframe
data.head()
```

```python
# Count the number of null values in the dataset
data.isnull().sum().sum()
```

```python
# Count the number of null values present in each column
data.isnull().sum()
```

* We need to impute the missing values in the Ratings , Location , Restro Type , Cuisines , Cost columns

<a id="imputation"><a>
## Missing values imputation

### Standard missing values

```python
# Impute null values in Location column with mode
data.Location.fillna(data.Location.mode().values[0],inplace=True)
# Check if null values have been imputed
data.Location.isnull().sum()
```

```python
# Impute null values in Restaurant type column with mode
data["Restro Type"].fillna(data["Restro Type"].mode().values[0],inplace=True)
# Check if null values have been imputed
data["Restro Type"].isnull().sum()
```

```python
# Impute null values in Cuisines column with mode
data.Cuisines.fillna(data.Cuisines.mode().values[0],inplace=True)
# Check if null values have been imputed
data.Cuisines.isnull().sum()
```

```python
# Convert datatype of Cost column from object to float by removing comma (,) 
data.Cost = data.Cost.apply(lambda a : a if str(a)=="nan" else float(str(a).replace(",","")))
# Impute null values in Cuisines column with mode
data.Cost.fillna(data.Cost.median(),inplace=True)
# Check if null values have been imputed
data.Cost.isnull().sum()
```

### Non-Standard missing values

```python
# Get unique records from Ratings column
data.Ratings.unique()
```

* Here, "NEW" and "-" are non standard missing values
* We need to convert them to null values

```python
# Convert non standard missing values to null
data.Ratings.replace({"-":np.nan,"NEW":np.nan},inplace=True)
# Convert datatype of rating from object to float
data.Ratings = data.Ratings.apply(lambda a : float(str(a).split("/")[0].replace(" ","")))
# Impute null values with median value
data.Ratings.fillna(data.Ratings.median(),inplace=True)
# Check if null values have been imputed
data.Ratings.isnull().sum()
```

```python
# Check if all the missing values have been imputed
data.isnull().sum()
```

#### 8.	Find the distribution of the votes and Approx_cost using a user-defined function and for a loop.

```python
# Retrieve the required columns individually from the dataset
votes = data[data['Cost'] != 0]['Votes'] # Extract values from Votes column
cost = data[data['Cost'] != 0]['Cost'] # Extract values from Cost column

# Define function to find the distribution
def find_distribution(data):
    
    # Initialize an empty dictionary
    distribution={} 
    
    # Use a for loop to find frequency of each of the unique values int the dataset
    for i in data:
        if i in distribution: # Condition to check if value is present in dictionary
            distribution[i]+=1 # Increament frequency value if the value is already present in the dictionary
        else:
            distribution[i]=1 # Add new item to the dictionary with key as data value and value as 1 if value is not present  
    
    # Return the dictionary which has all the unique values of the dataset and their respective frequencies        
    return distribution 

# Find distribution of votes and cost column using the user defined function
votes_dist = find_distribution(votes) # Votes distribution
cost_dist = find_distribution(cost)  # Cost distribution
```

```python
# Create dataframe for Cost distribution
df_cost = pd.DataFrame({"Cost":cost_dist.keys(),"Count":cost_dist.values()},dtype=int).sort_values("Cost")

# Adjust plot size for better visualization
plt.figure(figsize=(12,8))
# Create a bar plot with Cost in x-axis and frequency of each cost in y-axis
plt.bar(x=np.arange(len(df_cost.Cost)),height=df_c.Count,width=0.8)
# Assign respective cost values for xticks parameter
plt.xticks(np.arange(len(df_cost.Cost)),df_cost.Cost,rotation=90)

# Set various attributes for the plot
plt.title('Approximate cost for two people distribution',size=17,pad=15) # Set title for the plot
plt.xlabel('Cost',size = 15) # Set x-axis label for the plot
plt.ylabel('Count of restaurants',size = 15) # Set y-axis label for the plot

# Display the plot
plt.show()
```

* The distribution of cost is bell shaped and is positively skewed (skewed towards right)
* Around 1400 restaurants in Bangalore have cost of around Rs.300

```python
# Create dataframe for Vote distribution 
df_votes = pd.DataFrame({"Vote":votes_dist.keys(),"Count":votes_dist.values()},dtype=int).sort_values("Vote")
# Ignore the restaurants which have zero votes
df_votes.drop(index=df_votes[df_votes.Vote==0].index,inplace=True)

# Adjust plot size for better visualization
plt.figure(figsize=(12,8))
# Create a bar plot with Cost in x-axis and frequency of each cost in y-axis
plt.bar(x=np.arange(len(df_votes.Vote)),height=df_votes.Count,width=0.8)
# Assign respective cost values for xticks parameter
plt.xticks(np.arange(len(df_votes.Vote)),df_votes.Vote,rotation=90,color="w")
plt.tick_params(bottom=False)

# Set various attributes for the plot
plt.title('Votes distribution',size=17,pad=15) # Set title for the plot
plt.xlabel('Vote',size = 15) # Set x-axis label for the plot
plt.ylabel('Count of restaurants',size = 15) # Set y-axis label for the plot

# Display the plot
plt.show()
```

* Distribution of votes is highly skewed towards right.
* It is conclusive that many restaurants in Bangalore have minimum amount of votes.

#### 14.	Which are the most popular cuisines of Bangalore using a related plot?

```python
# Initialize empty list to store cuisines
cuisines=[]

# Use for loop to get all the row values in cuisines column from the dataset
for i in data.Cuisines:
    # Use inner for loop to get all the cuisines in a row which are separted by comma
    for j in i.split(","):
        cuisines.append(j.lstrip(" ")) # Append each and every cuisine to the list

# Find frequency for each cuisines and retrieve the top 20 records
popular_cuisines = pd.Series(cuisines).value_counts()[:20]

# Adjust plot size for better visualization
plt.figure(figsize=(10,5))
# Plot a bar chart for each cuisine and its total count 
plt.bar(x=popular_cuisines.index,height=popular_cuisines.values)
        
# Set various attributes of the plot
plt.title('Most Popular Cuisines of Bangaluru', size = 20, pad = 15) # Set title for the plot
plt.xlabel('Cuisines', size = 20) # Set x-axis label for the plot
plt.ylabel('Count', size = 20) # Set y-axis label for the plot
plt.xticks(size = 10, rotation = 90) # Set x-axis ticks size
plt.yticks(size = 10) # Set y-axis ticks size

# Display grid for the plot for better understanding
plt.grid()
# Display the plot
plt.show()
```
